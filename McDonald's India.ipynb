```python
li = []

for i in df.Per_Serve_Size:
    i= i.replace("ml", "").strip().replace("g" , "")
    li.append(i)
```

```python
df["Per_Serve_Size"] = li
```

```python
df.Per_Serve_Size= df.Per_Serve_Size.astype(float)
```

```python
df.Per_Serve_Size.dtypes
```

```python
size_data= pd.DataFrame(df.Per_Serve_Size.sort_values(ascending=False))
size_data["Menu_Items"] = df.Menu_Items

size_data[:10]
```

```python
plt.figure(figsize=(14 , 10))
plt.bar(x= size_data["Menu_Items"][:10] , height=size_data["Per_Serve_Size"][:10] , color= "#9C0000", ec= "black")

plt.title("Top 10 Items Size" , color= "#FF2D03" , fontstyle= "italic", fontsize=20)

plt.xlabel("Items Name" , color= "#FF2D03" , fontstyle= "italic", fontsize=20)
plt.ylabel("Size of Items" , color= "#FF2D03" , fontstyle= "italic", fontsize=20)

plt.xticks(rotation= 90)

plt.show()
```

```python
E_data= pd.DataFrame(df["Energy(kCal)"].sort_values(ascending=False))
E_data["Menu_Items"] = df.Menu_Items

E_data[:10]
```

```python
def plot(data, xvar, yvar):
    
    plt.figure(figsize=(14 , 10))
    plt.bar(x= data[xvar][:10] , height=data[yvar][:10] , color= "#9C0000", ec= "black")

    plt.title("Top 10 Menu Items Contains "+yvar , color= "#FF2D03" , fontstyle= "italic", fontsize=20)

    plt.xlabel("Items Name" , color= "#FF2D03" , fontstyle= "italic", fontsize=20)
    plt.ylabel(yvar+" of Items" , color= "#FF2D03" , fontstyle= "italic", fontsize=20)

    plt.xticks(rotation= 90)
```

```python
plot(E_data , "Menu_Items", "Energy(kCal)")
```
